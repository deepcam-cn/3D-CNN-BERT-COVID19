```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import shutil
from tqdm import tqdm
```

```python
covidx_dir = '/media/ubuntu/MyHDataStor2/datasets/COVID-19/ICCV-MIA/'
covidx_img_dir= covidx_dir 

print(covidx_dir)
print(covidx_img_dir)
```

```python
subsets = ['train','val']
classes = ["covid","non-covid"]

count = 0 

for subset in subsets:     

    for c in classes:
                
        print("subset = {}, class = {}".format(subset,c))
        scans = os.listdir(covidx_img_dir+"{}/{}/".format(subset,c) )
        #print(scans)

        pbar = tqdm(total = len(scans))
        for s in scans: 
            
            pbar.update()
            
            files = os.listdir(covidx_img_dir+"{}/{}/{}".format(subset,c,s) )
            print(files)
                        
            for f in files: 
                
                f_full = covidx_img_dir+"{}/{}/{}/{}".format(subset,c,s,f) 
                if f.startswith('._'):
                    continue 
                
                img = cv2.imread(f_full, cv2.IMREAD_UNCHANGED)
                #print(img.shape) 
                
                if img is None: 
                    print("file {} has problem".format(f_full))
                    input("dbg image file")
                    continue 
                
                if img.shape != (512,512):
                    
                    print("file {} shape = {}".format(f_full,img.shape))                     
                    img = cv2.resize(img,(512,512),cv2.INTER_LINEAR)
                    cv2.imwrite(f_full,img)
                    count += 1                     

print(count)                
```
